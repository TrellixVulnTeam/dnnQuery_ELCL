```python
import numpy as np

def renew19(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0] = tokens[16]
    newtokens[1:3] = ['A', 'as']
    newtokens[3:5] = tokens[4:6]
    newtokens[5:9] = tokens[0:4]
    newtokens[9:12] = ['and', 'B', 'as']
    #newtokens[12:14] = tokens[12:14]
    newtokens[14:] = tokens[8:12]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew15(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0:3] = tokens[12:]
    newtokens[3:7] = tokens[8:12]
    newtokens[7] = 'as'
    newtokens[8:10] = tokens[4:6]
    newtokens[10:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew14(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0:2] = tokens[12:]
    newtokens[2:6] = tokens[8:12]
    newtokens[6] = 'as'
    newtokens[7:9] = tokens[4:6]
    newtokens[9:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew13(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0] = tokens[10]
    newtokens[1:3] = ['A', 'as']
    newtokens[3:6] = tokens[0:3]
    newtokens[6:9] = ['and','B', 'as']
    newtokens[9:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew11(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[8:]
    newtokens[3:7] = tokens[0:4]
    newtokens[7] = 'and'
    newtokens[8:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew10(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[8:]
    newtokens[2:6] = tokens[0:4]
    newtokens[6] = 'and'
    newtokens[7:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew7(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[4:]
    newtokens[3:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew6(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[4:]
    newtokens[2:] = tokens[0:4]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew5(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0] = 'count'
    newtokens[1:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

len_search = dict()
f2 = open('./rand_test2.lo','w')
with open('./rand_test.lo') as f:
    for line in f:
        tokens = line.split()
        length = len(tokens)
        if length not in len_search:
            len_search[length] = []
        len_search[length].append(line)
        if length == 19:
            newline = renew19(line)
            f2.write(newline+'\n')
        elif length == 15:
            newline = renew15(line)
            f2.write(newline+'\n')
        elif length == 14:
            newline = renew14(line)
            f2.write(newline+'\n')
        elif length == 13:
            newline = renew13(line)
            f2.write(newline+'\n')
        elif length == 11:
            newline = renew11(line)
            f2.write(newline+'\n')
        elif length == 10:
            newline = renew10(line)
            f2.write(newline+'\n')
        elif length == 7:
            newline = renew7(line)
            f2.write(newline+'\n')
        elif length == 6:
            newline = renew6(line)
            f2.write(newline+'\n')
        elif length == 5:
            newline = renew5(line)
            f2.write(newline+'\n')
        else:
            f2.write(line)
f2.close()

for key in len_search.keys():
    value = len_search[key]
    print 'length = %d, total examples: %d' %(key, len(value))
```
