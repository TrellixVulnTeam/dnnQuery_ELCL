```python
import numpy as np

def renew19(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0] = tokens[16]
    newtokens[1:3] = ['A', 'as']
    newtokens[3:5] = tokens[4:6]
    newtokens[5:9] = tokens[0:4]
    newtokens[9:12] = ['and', 'B', 'as']
    #newtokens[12:14] = tokens[12:14]
    newtokens[14:] = tokens[8:12]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew15(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0:3] = tokens[12:]
    newtokens[3:7] = tokens[8:12]
    newtokens[7] = 'as'
    newtokens[8:10] = tokens[4:6]
    newtokens[10:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew14(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0:2] = tokens[12:]
    newtokens[2:6] = tokens[8:12]
    newtokens[6] = 'as'
    newtokens[7:9] = tokens[4:6]
    newtokens[9:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew13(line):
    tokens = line.split()
    newtokens = [x for x in tokens[:-1]]
    newtokens[0] = tokens[10]
    newtokens[1:3] = ['A', 'as']
    newtokens[3:6] = tokens[0:3]
    newtokens[6:9] = ['and','B', 'as']
    newtokens[9:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew11(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[8:]
    newtokens[3:7] = tokens[0:4]
    newtokens[7] = 'and'
    newtokens[8:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew10(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[8:]
    newtokens[2:6] = tokens[0:4]
    newtokens[6] = 'and'
    newtokens[7:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew7(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[4:]
    newtokens[3:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

def renew6(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[4:]
    newtokens[2:] = tokens[0:4]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew5(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0] = 'count'
    newtokens[1:] = tokens[0:4]
    newline = ' '.join(newtokens)
    return newline

len_search = dict()
f2 = open('./rand_test2.lo','w')
with open('./rand_test.lo') as f:
    for line in f:
        tokens = line.split()
        length = len(tokens)
        if length not in len_search:
            len_search[length] = []
        len_search[length].append(line)
        if length == 19:
            newline = renew19(line)
            f2.write(newline+'\n')
        elif length == 15:
            newline = renew15(line)
            f2.write(newline+'\n')
        elif length == 14:
            newline = renew14(line)
            f2.write(newline+'\n')
        elif length == 13:
            newline = renew13(line)
            f2.write(newline+'\n')
        elif length == 11:
            newline = renew11(line)
            f2.write(newline+'\n')
        elif length == 10:
            newline = renew10(line)
            f2.write(newline+'\n')
        elif length == 7:
            newline = renew7(line)
            f2.write(newline+'\n')
        elif length == 6:
            newline = renew6(line)
            f2.write(newline+'\n')
        elif length == 5:
            newline = renew5(line)
            f2.write(newline+'\n')
        else:
            f2.write(line)
f2.close()

for key in len_search.keys():
    value = len_search[key]
    print 'length = %d, total examples: %d' %(key, len(value))
```

## Grammar Version 3.0
Used for generate new version of grammar (shorter version for sum, diff, avg)

```python
def renew19(line):
    ''' from 19 to 11
    '''
    tokens = line.split()
    # newly changed
    # 'sum {comp_field1} where {comp_field2} {comp} {comp_val} and where {comp_field2} {comp} {query2_comp_val} '
    newtokens = [x for x in tokens[:11]]
    newtokens[0] = tokens[16]
    newtokens[1] = tokens[5]
    newtokens[2:6] = tokens[0:4]
    newtokens[6] = 'and'
    newtokens[7] = 'where'
    #newtokens[12:14] = tokens[12:14]
    newtokens[8:11] = tokens[9:12]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew15(line, query):
    ''' case 1: nested type: from 15 to 11
        case 2: before/after: from 15 to 6
    '''
    tokens = line.split()
    words = query.split()
    if words[0] == 'of':
        # case 1
        newtokens = [x for x in tokens[:11]]
        newtokens[0] = 'select'
        newtokens[1] = tokens[13]
        newtokens[2] = tokens[12]
        newtokens[3] = tokens[14]
        newtokens[4:7] = tokens[8:11]
        newtokens[7:11] = tokens[0:4]
    else:
        # case 2
        newtokens = [x for x in tokens[:6]]
        newtokens[0] = 'select'
        newtokens[1] = tokens[13]
        if tokens[10] == 'greater':
            newtokens[2] = 'next'
        else:
            newtokens[2] = 'prev'    
        newtokens[3:6] = tokens[1:4]
    
    newline = ' '.join(newtokens)
    return newline

def renew14(line):
    ''' before/after type, from 14 to 6
        New grammar: select {field1} next/prev {field2} equal {value2}
    '''
    tokens = line.split()
    # newly change
    newtokens = [x for x in tokens[:6]]
    newtokens[0:2] = tokens[12:]
    if tokens[10] == 'greater':
        newtokens[2] = 'next'
    else:
        newtokens[2] = 'prev'    
    newtokens[3:6] = tokens[1:4]
    newline = ' '.join(newtokens)
    return newline

def renew13(line):
    ''' from 13 to 7
    '''
    tokens = line.split()
    # newly changed: diff {comp_field1}' arg {max_min} {arg1} 'and' arg {max_min} {arg1}
    newtokens = [x for x in tokens[:7]]
    newtokens[0] = tokens[10]
    newtokens[1] = tokens[1]
    newtokens[2] = tokens[0]
    newtokens[3] = tokens[2]
    newtokens[4] = 'and'
    newtokens[5] = tokens[5]
    newtokens[6] = tokens[7]
    newline = ' '.join(newtokens)
    return newline

def renew11(line):
    ''' from 11 to 12
    '''
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[8:]
    newtokens[3:7] = tokens[0:4]
    newtokens[7] = 'and'
    newtokens[8:] = tokens[5:8]
    
    # newly added
    newtokens.insert(0, 'select')
    temp = newtokens[1]
    newtokens[1] = newtokens[2]
    newtokens[2] = temp
    newline = ' '.join(newtokens)
    return newline

def renew10(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[8:]
    newtokens[2:6] = tokens[0:4]
    newtokens[6] = 'and'
    newtokens[7:] = tokens[5:8]
    newline = ' '.join(newtokens)
    return newline

def renew7(line):
    ''' from 7 to 8
    '''
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:3] = tokens[4:]
    newtokens[3:] = tokens[0:4]
    
    # newly added
    newtokens.insert(0, 'select')
    temp = newtokens[1]
    newtokens[1] = newtokens[2]
    newtokens[2] = temp
    newline = ' '.join(newtokens)
    return newline

def renew6(line):
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0:2] = tokens[4:]
    newtokens[2:] = tokens[0:4]
    
    if newtokens[0] == 'mean':
        newtokens[0] = 'avg'
    newline = ' '.join(newtokens)
    return newline

def renew5(line, schema):
    ''' from 5 to 6
    '''
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens[0] = 'count'
    newtokens[1:] = tokens[0:4]
    
    # newly added
    fields = schema.split()
    newtokens.insert(1, fields[0])
    newline = ' '.join(newtokens)
    return newline

def renew3(line):
    ''' from 3 to 4
    '''
    tokens = line.split()
    newtokens = [x for x in tokens]
    newtokens.insert(0, 'select')
    temp = newtokens[1]
    newtokens[1] = newtokens[2]
    newtokens[2] = temp
    newline = ' '.join(newtokens)
    return newline
```
