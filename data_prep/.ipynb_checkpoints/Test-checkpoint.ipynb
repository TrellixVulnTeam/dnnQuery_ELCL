We use Stanford Parser (NLTK python interface) to get the dependency tree for each sentence, later we will analyze the tree

```python
import os
from nltk.parse import stanford
from nltk import tree as nt

os.environ['STANFORD_PARSER'] = '/Users/richard_xiong/Documents/DeepLearningMaster/jars'
os.environ['STANFORD_MODELS'] = '/Users/richard_xiong/Documents/DeepLearningMaster/jars'
parser = stanford.StanfordParser(model_path='/Users/richard_xiong/Documents/DeepLearningMaster/jars/englishPCFG.ser.gz')

# the 0th sentence is there to ensure the functioning
sentences = parser.raw_parse_sents(('Hello, My name is Melroy',"when the 1st_venue was beijing and 2nd_venue was dubai, which city was the most recent 3rd_venue"))
#print sentences
print sentences[1]   # Tree object
print list(sentences[1])

dep = sentences[1].leaves()
print dep
# for line in sentences[1]:
#     line.draw()
```

In order to find which value corresponds to which field, we need to first find:
1. the lowest common ancestor for each (value, field) pairs
2. for each value, all different ancestors are belong to different levels, the deepest one, which should be the subtree for all the others, would contain the correspondence pair

Possible functions:

leaf_treeposition(self, index) ---> return: The tree position of the ``index``-th leaf in this
            tree.  I.e., if ``tp=self.leaf_treeposition(i)``, then
            ``self[tp]==self.leaves()[i]``.

treeposition_spanning_leaves(self, start, end) ---> The tree position of the lowest descendant of this
            tree that dominates ``self.leaves()[start:end]``.

convert(cls, tree) ---> to subtype of Tree, say, ParentTree

```python
print sentences[1].leaf_treeposition(2)
print sentences[1].leaf_treeposition(4)
print sentences[1].leaf_treeposition(6)
print sentences[1].leaf_treeposition(8)

print sentences[1].treeposition_spanning_leaves(2, 4) 
print sentences[1].treeposition_spanning_leaves(4, 6) 
```

```python
schema = 'Player Matches Innings Runs Average 100s 50s Games_Played Field_Goals Free_Throws Points'
fields = schema.split()

# how many players played more than 10 matches
# where Matches greater 10 sum Player
for i in range(1,4):
    print 'how many players played more than 30 '+fields[i].lower()
    print 'how many players played less than 30 '+fields[i].lower()

print '\n'

for i in range(1,4):
    print 'where ' +fields[i] +' greater 30 sum Player'
    print 'where ' +fields[i] +' less 30 sum Player'
    
for i in range(18):
    print schema
```

```python
schema = 'Discipline Amanda Bernie Javine_H Julia Michelle'
fields = schema.split()

options = 'Hurdles Cycling Swimming Curling Archery Hammer'
values = options.split()
# what are the number of points bernie scored in hurdles
# where Discipline equal Hurdles select Bernie
for i in range(1,6):
    for j in range(6):
        print 'what is the number of points '+ fields[i].lower()+' scored in '+values[j].lower()
    
print '\n'

for i in range(1,6):
    for j in range(6):
        print 'where Discipline equal ' +values[j] +' select '+fields[i]
        
print '\n'

for i in range(30):
    print schema
```

```python
schema = 'Player Matches Innings Runs Average 100s 50s Games_Played Field_Goals Free_Throws Points'
fields = schema.split()

options = 'John Albert Bill Herbie George'
values = options.split()
# how many more runs does john have than albert
# where Player equal John select Runs as A where Player equal Albert select Runs as B diff A B

for i in range(1,6):
    for j in range(5):
        for k in range(5):
            if j == k:
                continue
            print 'how many more '+ fields[i].lower()+' does '+values[j].lower()+' have than '+values[k].lower()
    
print '\n'

for i in range(1,6):
    for j in range(5):
        for k in range(5):
            if j == k:
                continue
            print 'where Player equal ' +values[j] +' select '+fields[i]+' as A where Player equal '\
            +values[k] +' select '+fields[i]+' as B diff A B'
        
print '\n'

for i in range(100):
    print schema
```

```python
schema = 'Swara Position Short_name Notation Mnemonic'
fields = schema.split()

options = 'Shatshruti_Dhaivata Shadja Panchama Antara_Gandhara'
values = options.split()
# what swara is above shatshruti_dhaivata
# where Swara equal Shatshruti_Dhaivata select Position as A where Position less A argmax Swara Position

for i in range(4):
    print 'what swara is above '+values[i].lower()
    print 'what swara is below '+values[i].lower()

print '\n'

for i in range(4):
    print 'where Swara equal ' +values[i] +' select Position as A where Position less A argmax Swara Position'
    print 'where Swara equal ' +values[i] +' select Position as A where Position greater A argmin Swara Position'
    
print '\n'

for i in range(8):
    print schema
```
